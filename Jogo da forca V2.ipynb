```python
import random
from os import system, name

# Função para limpar a tela a cada execução do jogo
def limpa():
    # Windowns
    if name == 'nt':
        _= system('cls')
    # Mac ou Linux
    else:
        _= system('clear')
    

# Função para escolher o tipo de jogo 
def opcao():
    
    print("\nBem vindo ao jogo da forca!")
    print("Escolha o tipo de operação:\n\n 1 - Fruta\n 2 - Animais\n 3 - Cores\n 4 - Esportes\n")

    # Variável para armazenar a opção do jogo
    opcao = input("Digite sua opção: ")

    if opcao == '1':
        jogoFrutas()
        
    elif opcao == '2':
        jogoAnimais()
        
    elif opcao == '3':
        jogoCores()
        
    elif opcao == '4':
        jogoEsportes()
        
    else: 
        print("\nEscolha inválida. Tente novamente.\n")


# Função que desenha a forca na tela
def desenho(chances):

    # Lista de estágios da forca
    stages = [  # estágio 6 (final)
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """,
                # estágio 5
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                """,
                # estágio 4
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                """,
                # estágio 3
                """
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                """,
                # estágio 2
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                # estágio 1
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                # estágio 0
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return stages[chances]


# Função do jogo
def jogo(categoria, palavras):
    limpa()
    print(f"\nAdivinhe a palavra abaixo:\n")
    
    # Escolhe randomicamente uma palavra
    palavra = random.choice(palavras)

    # Lista de letras da palavra
    lista_letras = list(palavra)

    # Cria o tabuleiro com o caracter "_" multiplicado pelo comprimento da palavra
    tabuleiro = ["_"] * len(palavra)

    chances = 6
    letras_tentativas = []

    # Loop enquanto número de chances for maior do que zero
    while chances > 0:

        # Mostra o desenho no primeiro estágio
        print(desenho(chances))
        print(f"{categoria}: ", " ".join(tabuleiro))

        # Pedindo ao usuário para digitar uma letra e transformando em minúscula caso ele digite em maiúscula
        tentativa = input("\nDigite uma letra: ").lower()

        # Condicional se a tentativa estiver sendo repetida
        if tentativa in letras_tentativas:
            print("Você já tentou essa letra. Escolha outra!")
            continue

        # Lista de tentativas (armazenando as letras)
        letras_tentativas.append(tentativa)

        # Condicional se a tentativa estiver na palavra
        if tentativa in lista_letras:
            print("Você acertou a letra!")

            # Loop para cada indice percorrendo a fruta
            for i, letra in enumerate(lista_letras):
                if letra == tentativa:
                    tabuleiro[i] = tentativa

            # Se todos os espaços foram preenchidos, o jogo acabou
            if "_" not in tabuleiro:
                print(f"\nVocê venceu! A palavra era: {palavra}\n")
                return

        # Se não, as chances vão se diminuindo a cada erro 
        else:
            print("Ops. Essa letra não está na palavra!")
            chances -= 1

    # Condicional se ainda estiver letra faltando quando as chances passarem de 6
    if "_" in tabuleiro:
        print(f"\nVocê perdeu! A palavra era: {palavra}.\n")


# Função do jogo para fruta
def jogoFrutas():
    frutas = ['banana', 'abacate', 'uva', 'morango', 'laranja', 'pera', 'abacaxi', 'goiaba']
    jogo('fruta', frutas)


# Função do jogo para animais
def jogoAnimais():
    animais = ['macaco', 'tigre', 'girafa', 'elefante', 'leopardo', 'urso', 'tartaruga']
    jogo('animal', animais)


# Função do jogo para cores
def jogoCores():
    cores = ['azul', 'vermelho', 'amarelo', 'rosa', 'verde', 'vermelho', 'branco', 'preto']
    jogo('cor', cores)


# Função do jogo para esporte
def jogoEsportes():
    esportes = ['volei', 'basquete', 'futebol', 'hipismo', 'ginástica', 'natação']
    jogo('esporte', esportes)


# Inicia o jogo
if __name__ == "__main__":
    opcao()
```

# Jogo da forca

```python
import random
from os import system, name

# Função para limpar a tela a cada execução do jogo
def limpa():
    # Windowns
    if name == 'nt':
        _= system('cls')
    # Mac ou Linux
    else:
        _= system('clear')
      
# Função que desenha a forca na tela
def desenho(chances):

    # Lista de estágios da forca
    stages = [  # estágio 6 (final)
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / \\
                   -
                """,
                # estágio 5
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |     / 
                   -
                """,
                # estágio 4
                """
                   --------
                   |      |
                   |      O
                   |     \\|/
                   |      |
                   |      
                   -
                """,
                # estágio 3
                """
                   --------
                   |      |
                   |      O
                   |     \\|
                   |      |
                   |     
                   -
                """,
                # estágio 2
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                # estágio 1
                """
                   --------
                   |      |
                   |      O
                   |    
                   |      
                   |     
                   -
                """,
                # estágio 0
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return stages[chances]

# Função do jogo
def jogo():

    limpa()
    print("\nBem vindo ao jogo da forca!")
    print("Adivinhe a fruta abaixo:\n")
    
    # Lista de frutas
    frutas = ['banana', 'abacate', 'uva', 'morango', 'laranja']
    
    # Escolhe randomicamente uma fruta
    fruta = random.choice(frutas)
    
    # Lista de letras da fruta
    lista_letras_frutas = [letra for letra in fruta]
    
    # Cria o tabuleiro com o caracter "_" multiplicado pelo comprimento da fruta
    tabuleiro = ["_"] * len(fruta)
    
    # Número de chances
    chances = 6
    
    # Lista para as letras digitadas
    letras_tentativas = []
    
    # Loop enquanto número de chances for maior do que zero
    while chances > 0:
        
        # Mostra o desenho no primeiro estágio
        print(desenho(chances))
        print("Fruta: ", tabuleiro)
        print("\n")
        
        # Tentativa
        tentativa = input("\nDigite uma letra: ")
        
        # Condicional se a tentativa estiver sendo repetida
        if tentativa in letras_tentativas:
            print("Você já tentou essa letra. Escolha outra!")
            continue
        
        # Lista de tentativas (armazenando as letras)
        letras_tentativas.append(tentativa)
        
        # Condicional se a tentativa estiver na fruta
        if tentativa in lista_letras_frutas:
            
            print("Você acertou a letra!")
            
            # Loop para cada indice percorrendo a fruta
            for indice in range(len(lista_letras_frutas)):

                # Condicional se o indice da letra for igual a tentativa
                if lista_letras_frutas[indice] == tentativa:
                    # Mostra a letra no seu indice 
                    tabuleiro[indice] = tentativa
            
            # Se todos os espaços foram preenchidos, o jogo acabou
            if "_" not in tabuleiro:
                print("\nVocê venceu! A fruta era: {}".format(fruta))
                break
        else:
            # Se não, as chances vão se diminuindo a cada erro 
            print("Ops. Essa letra não está na fruta!")
            # Decremento
            chances -= 1
    
    # Condicional se ainda estiver letra faltando quando as chances passarem de 6
    if "_" in tabuleiro:
        print("\nVocê perdeu! A fruta era: {}.".format(fruta))


# Bloco main
if __name__ == "__main__":
    jogo()
    print("\nParabéns!\n")
```
